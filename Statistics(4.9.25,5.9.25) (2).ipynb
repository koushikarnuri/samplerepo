### Z-test

##### 1.	The average height of all residents in a city is 168cm with a population std = 3.9. A doctor believes the mean is different. He measured the height of 36 individuals and found the average to be 169.5cm 
(a). State Null and Alternative hypothesis
(b) At a 95% ,CI is there enough evidence to reject the Null Hypothesis.


```python
#### Given_data
pop_mean=168
pop_std_dev=3.9
n=36
x_bar=169.5
### CI--95% (0.95), alpha-5%(0.05)
```

```python
### H0--Pop mean is same(Pop_mean==168)
### Ha--Pop mean is different(Pop_mean!=168)
```

```python
zscore= (x_bar-pop_mean)/(pop_std_dev/np.sqrt(n))
zscore
```

```python
### from z_table , z_critical =1.96
### Since z_critical < z_score, reject H0
### Pop mean is different(Pop_mean!=168)
```

```python
s=pop_std_dev/np.sqrt(n)
s
```

```python
sample_data= stats.norm.rvs(size=n,loc=x_bar,scale=s)
sample_data
```

```python
z_critical=stats.norm.ppf(0.975)
z_critical
```

```python
from statsmodels.stats.weightstats import ztest
zscore,p_value=ztest(x1=sample_data,value=pop_mean)
print(zscore)
```

## 2.	Scores on an exam are normally distributed with a population standard deviation=5.6. a random sample of 40 scores on the exam has a mean of 32.
Estimate the population mean with
a)	80% confidence 
b)	90% confidence
c)	98 confidences


```python
### Given data
pop_std_dev=5.6
x_bar=32
n=40
```

```python
### a) 80% confidence
z_critical_80= stats.norm.ppf(0.90)
z_critical_80
```

```python
pop_mean_lower= x_bar-(z_critical_80*(pop_std_dev/np.sqrt(n)))
pop_mean_lower
```

```python
pop_mean_upper= x_bar+(z_critical_80*(pop_std_dev/np.sqrt(n)))
pop_mean_upper
```

```python
### b) 90% confidence
z_critical_90= stats.norm.ppf(0.95)
z_critical_90
```

```python
pop_mean_lower= x_bar-(z_critical_90*(pop_std_dev/np.sqrt(n)))
pop_mean_lower
```

```python
pop_mean_upper= x_bar+(z_critical_90*(pop_std_dev/np.sqrt(n)))
pop_mean_upper
```

```python
### c) 98% confidence
z_critical_98= stats.norm.ppf(0.99)
z_critical_98
```

```python
pop_mean_lower= x_bar-(z_critical_98*(pop_std_dev/np.sqrt(n)))
pop_mean_lower
```

```python
pop_mean_upper= x_bar+(z_critical_98*(pop_std_dev/np.sqrt(n)))
pop_mean_upper
```

#### 3.	In the population the average IQ is 100. A team of scientists wants to test a new medication to see if it has either a positive or negative effect on intelligence, or no effect at all. A sample of 30 participants who have been taken the medications has a mean of 140 with a standard deviation of 20. Did the medication affect the intelligence. alpha=0.05.

```python
### H0--Medication does not affect the Intelligence
### Ha-- Medication affects the Intelligence
### Given data
pop_mean=100
n=30
x_bar=140
s=20
### alpha=0.05, CI=0.95
```

```python
z_score= (x_bar-pop_mean)/s
z_score
```

```python
z_critical= stats.norm.ppf(0.975)
z_critical
```

```python
### Since z_critical < z_stats, reject H0
### Medication affects the Intelligence
```

## 4.	A random sample of 27 observations from a large population has a mean of 22 and a standard deviation of 4.8. Can we conclude at alpha=0.01 that the population mean is significantly below 24?

```python
### H0--Pop mean is above 24
### Ha--Pop mean is below 24
```

```python
### Given data
n=27
x_bar=22
s=4.8
pop_mean=24
## alpha=0.01, CI=0.99
```

```python
tstats= (x_bar-pop_mean)/(s/np.sqrt(n))
tstats
```

```python
t_critical= stats.t.ppf(0.005,26)
t_critical
```

```python
sample_data= stats.norm.rvs(size=n,loc=x_bar,scale=s)
sample_data
```

```python
stats.ttest_1samp(sample_data,popmean=pop_mean)
```

```python
### Since t_critical < t_stats in -ve region , fail to reject H0
### Pop mean is above 24
```

#### 5.	A random sample of 15 observation has a mean of 20 and standard deviation of 3.5. To estimate population mean with 95% confidence level and find the margin of error.

```python
### Given data
n=15
x_bar=20
s=3.5
### CI=95%, alpha=5%
```

```python
t_critical= stats.t.ppf(0.975,14)
t_critical
```

```python
Margin_of_Error= t_critical*(s/np.sqrt(n))
Margin_of_Error
```

```python
pop_mean_lower= x_bar-Margin_of_Error
pop_mean_upper= x_bar+Margin_of_Error
print(pop_mean_lower,',',pop_mean_upper)
```

### 6.	A researcher wants to compare the effectiveness of two different medications for reducing blood pressure. Medication A is tested on 50 patients, resulting in a mean reduction of 15 mmHg with a standard deviation of 3 mmHg. Medication B is tested on 60 patients, resulting in a mean reduction of 13 mmHg with a standard deviation of 4 mmHg. At a 1% significance level, is there a significant difference between the two medications?

```python
### Given data
### Sample A
n1=50
x1_bar=15
s1=3
### Sample B
n2= 60
x2_bar=13
s2=4
### alpha=1% ,CI--99%
## H0--No significance difference between the two medications.
## Ha--Significance difference between the two medications.
```

```python
zscore= (x1_bar-x2_bar)/np.sqrt(((s1**2)/n1) +((s2**2)/n2))
zscore
```

```python
z_critical = stats.norm.ppf(1-(0.01/2))
z_critical
```

```python
### Since z_critical < z_stats reject H0
### Significance difference between the two medications.
```
