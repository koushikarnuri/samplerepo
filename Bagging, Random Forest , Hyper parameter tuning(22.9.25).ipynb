```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler,OrdinalEncoder,LabelEncoder
import warnings
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('sales.csv',index_col=[0])
```

```python
df.shape
```

```python
df.head()
```

```python
#### Missing values
df.isnull().sum()
```

```python
### duplicates
df.duplicated().sum()
```

```python
target=df[['high']]
features= df.drop(columns=['high'])
features.columns
```

```python
ord_enc=OrdinalEncoder()
features[[ 'ShelveLoc','Urban', 'US']]= ord_enc.fit_transform(features[[ 'ShelveLoc','Urban', 'US']])
features.head()
```

```python
plt.figure(figsize=(10,7))
features.boxplot()
plt.show()
```

```python
### outlier capping
def outlier_capping(df,column):
    Q1= df[column].quantile(0.25)
    Q3= df[column].quantile(0.75)
    IQR= Q3-Q1
    lower_extreme= Q1-1.5*IQR
    upper_extreme= Q3+1.5*IQR
    df[column]= df[column].apply(lambda x: lower_extreme if x<lower_extreme else upper_extreme if x>upper_extreme else x)
for col in features.select_dtypes(['int','float']).columns:
    outlier_capping(features,col)
```

```python
features.boxplot()
plt.show()
```

```python
target.head()
target1=target.copy()
le= LabelEncoder()
target1=pd.DataFrame(le.fit_transform(target1),columns=target.columns)
target1.head()
```

```python
features.head()
```

```python
### feature selection
from sklearn.feature_selection import f_classif
f_class= f_classif(features,target1)
pd.Series(f_class[0],index=features.columns).sort_values(ascending=False).plot(kind='bar')
plt.show()
```

```python
features.columns
```

```python
features.drop(columns=['Population','Education'],inplace=True)
features.head()
```

```python
x_train,x_test,y_train,y_test= train_test_split(features,target1,train_size=0.7,random_state=150,stratify=target1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
std_sca=StandardScaler()
x_train[['CompPrice', 'Income', 'Advertising', 'Price','Age']]=std_sca.fit_transform(x_train[['CompPrice', 'Income', 'Advertising', 'Price','Age']])
x_test[['CompPrice', 'Income', 'Advertising', 'Price','Age']]= std_sca.transform(x_test[['CompPrice', 'Income', 'Advertising', 'Price','Age']])
```

```python
x_train.columns
```

##### Bagging Classifier

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
bag_class= BaggingClassifier(estimator=DecisionTreeClassifier(),n_estimators=100,bootstrap=True,
                             max_features=1,bootstrap_features=True,max_samples=0.9,random_state=50)
```

```python
bag_class.fit(x_train,y_train)
y_pred= bag_class.predict(x_test)
accuracy_score(y_test,y_pred)
```

#### RandomForest

```python
rand_for= RandomForestClassifier(n_estimators=100,bootstrap=True,max_features='sqrt',max_samples=0.9,
                                random_state=50)
```

```python
rand_for.fit(x_train,y_train)
y_pred=rand_for.predict(x_test)
accuracy_score(y_test,y_pred)
```

##### Hyper_parameter

```python
params={'n_estimators':[50,100,150,200],'max_features':[0.5,0.6,0.7,0.8,0.9,1],'max_samples':[0.7,0.8,0.9,1]}
```

```python
grid_search=GridSearchCV(bag_class,params,cv=5)
grid_search.fit(x_train,y_train)
grid_search.best_params_
```
