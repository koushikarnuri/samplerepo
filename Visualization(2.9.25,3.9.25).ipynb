```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
Exp=[1,2,4,5,6]
Salary=[1000,2000,4000,5000,10000]
```

```python
import matplotlib.colors as mc
mc.CSS4_COLORS
```

```python
import matplotlib.font_manager as mf
mf.fontManager.get_font_names()
```

```python
## Line chart
plt.plot(Exp,Salary,color='red',marker='*',markerfacecolor='beige',markersize=10,linewidth=2,linestyle=':')
plt.xlabel('Experience',fontname='Tahoma',fontsize=10,color='red')
plt.ylabel('Salary',fontsize=10, color='red')
plt.title('Exp Vs Salary',fontname='Tahoma',fontsize=12, color='blue');
```

```python
### Scatter plot
plt.scatter(Exp,Salary,color='red',linewidth=4)
```

```python
### barchart
a=plt.bar(Exp,Salary,color='red',edgecolor='black',width=0.8)
plt.bar_label(a);
```

```python
### Pie chart
Marks=[90,100,80,75,60]
Names=['Ajay','Sanjay','Suman','Ramya','John']
colors= ['red','blue','brown','green','orange']
```

```python
plt.pie(Marks,labels=Names,colors=colors,radius=1,autopct='%.2f%%',textprops={'color':'black','size':10},
       wedgeprops={'edgecolor':'black','linewidth':3},explode=[0,0.1,0,0,0],shadow=True);
```

```python
print(plt.style.available)
```

```python
### plotting subplots
fig,axs= plt.subplots(2,2,figsize=(12,6))
fig.suptitle('Peformance Measures',size=12,color='green',fontweight='bold',y=1)
axs[0,0].plot(Exp,Salary,color='red',marker='*',markerfacecolor='beige',markersize=10,linewidth=2,linestyle=':')
axs[0,0].set_xlabel('Exp',size=10,color='red')
axs[0,0].set_ylabel('Sal',size=10,color='red')
axs[0,0].set_title('Exp vs Sal',color='red',size=10)
axs[0,1].scatter(Exp,Salary,color='green')
axs[0,1].set_xlabel('Exp',size=10,color='red')
axs[0,1].set_ylabel('Sal',size=10,color='red')
axs[0,1].set_title('Exp vs Sal',color='red',size=10)
axs[1,0].pie(Marks,labels=Names,colors=colors,radius=1,autopct='%.2f%%',textprops={'color':'black','size':10},
       wedgeprops={'edgecolor':'black','linewidth':3},explode=[0,0.1,0,0,0],shadow=True)
axs[1,1].bar(Exp,Salary,color='red',edgecolor='black',width=0.8)
axs[1,1].set_xlabel('Exp',size=10,color='red')
axs[1,1].set_ylabel('Sal',size=10,color='red')
axs[1,1].set_title('Exp vs Sal',color='red',size=10)
plt.tight_layout();
```
