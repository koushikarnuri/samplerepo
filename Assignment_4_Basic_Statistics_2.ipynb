Problem Statement : Hospital Patient Data Analysis

```python
import numpy as np
import pandas as pd
```

1. Load the patient dataset and show summary with info().

```python
df=pd.read_csv('/content/Patient_Data.csv')
df1=pd.read_csv('/content/Billing_Data.csv')
```

```python
df.info()
```

2.	Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount']

```python
pd.DataFrame(df,columns=['PatientID', 'Department', 'Doctor', 'BillAmount'])
```

3.	Drop administrative columns like ['ReceptionistID', 'CheckInTime'].

```python
df.drop(['ReceptionistID', 'CheckInTime'],axis=1,inplace=True)
```

```python
df
```

4.	Use groupby to find total bill amount per department.

```python
df.groupby('Department')['BillAmount'].sum()
```

5.	Remove duplicate patient records based on PatientID.

```python
df.drop_duplicates(subset='PatientID',inplace=True)
df
```

6.	Fill missing BillAmount values with the mean bill amount.

```python
df.BillAmount.fillna(df.BillAmount.mean(),inplace=True)
df
```

7.	Merge the billing dataset with patient dataset on PatientID.

```python
a=df1.merge(df,on='PatientID')
a
```

8.	Concatenate an additional DataFrame that contains new patients for the current week (row-wise).

```python
new_df = pd.DataFrame({'PatientID': [106, 107],'Department': ['Cardiology', 'Neurology'],'Doctor': ['Dr. Smith', 'Dr. Allen'],'BillAmount': [1500, 2000]})
b= pd.concat([a, new_df], axis=0, ignore_index=True)
b
```

9.	Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).

```python
new_col=pd.DataFrame({'InsuranceCovered':[1000,1500],'FinalAmount':[1500,2000]})
c=pd.concat([b.reset_index(drop=True),new_col.reset_index(drop=True)], axis=1)
c
```
