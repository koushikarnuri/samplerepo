**Handling Null Values**

```python
import pandas as pd
import numpy as np
data = [
    [1,2,3,'M'],
    [np.nan,5,6,np.nan],
    [7,np.nan,8,'F'],
    [np.nan,np.nan,np.nan,'F']
]
```

```python
dframe = pd.DataFrame(data,columns=['a','b','c','d'])
dframe
```

```python
dframe['a']
```

```python
dframe.fillna(0)
```

```python
dframe
```

```python
# find avg of column a. It is neglecting null values from that column
avg_a = dframe['a'].mean()
avg_a
```

```python
avg_b = dframe['b'].mean()
avg_b
```

```python
avg_c = round(dframe['c'].mean(),1)
#avg_c = dframe['c'].mean()
avg_c
```

```python
dframe.shape[1]
```

```python
# Change null values of categorical column by finding mode of column
dframe['d'].mode() # we get
```

```python
d_mode = dframe['d'].mode()[0]
d_mode
# if bimodal or trimodal data, you will get 2 or 3 values as mode. So you can give [index] of mode to select mode out of 2 or 3
```

```python
# to change original dataframe use inplace=True
dframe.fillna({'a':avg_a,'b':avg_b,'c':avg_c,'d':d_mode},inplace=True)
```

```python
dframe
```

```python
# To retain your original dataframe and changed dataframe
# Keep original in one dataframe and changed in another dataframe without using inplace=True

# Original Dataframe
dframe = pd.DataFrame(data,columns=['a','b','c','d'])
dframe
```

```python
# Assign changed dataframe to new dataframe
newdf = dframe.fillna({'a':avg_a,'b':avg_b,'c':avg_c,'d':d_mode})
newdf
```

```python
dframe
```

```python
# remove null values
dframe.dropna() # original dataframe is not changed
```

```python
dframe
```

**Lambda Function**
A lambda function in Python is a small anonymous function — meaning it doesn’t have a name (unless you assign it one) and is typically used for short, simple operations.
Syntax: **lambda arguments: expression**

*   Suppose we want to write a function which will look at the years of experience.
*   If it is more than 20 years we will say "Very Experienced" and if less than 20, we will say "Less Experienced". For that let us write lambda function.



```python
fn = lambda x : 'Very Experienced' if x > 20 else 'Less Experienced'
```

```python
fn(30)
```

```python
fn(10)
```

```python
sal
```

Suppose we want to apply this function on each record of our dataframe i.e. on Service column. For that use apply() function. It is a row level function.

```python
sal['service'].apply(lambda x : 'Very Experienced' if x > 20 else 'Less Experienced')
```

```python
sal
```

To add this column in your existing dataframe

```python
sal['status'] = sal['service'].apply(lambda x : 'Very Experienced' if x > 20 else 'Less Experienced')
```

**datetime module**

```python
from datetime import date # date related functions
```

```python
today = date.today()
print("Today's date is: ",today)
```

```python
# Current date in different formats
# dd/mm/YY
d1 = today.strftime("%d/%m/%Y") # dd/mm/yyyy 27/08/2024
print("d1 =", d1)

# Textual month, day and year
d2 = today.strftime("%B %d, %Y") # August 27, 2024, %B-Full month name, %Y-yyyy, %y-yy
print("d2 =", d2)

# mm/dd/yy
d3 = today.strftime("%m/%d/%y") # mm/dd/yy 08/27/24
print("d3 =", d3)

# Month abbreviation, day and year
d4 = today.strftime("%b-%d-%Y") # Dec-15-2023, %b-abb. of monthname
print("d4 =", d4)
```

```python
from datetime import datetime # date and time related functions
now = datetime.now()

print("now =", now)

# dd/mm/YY H:M:S
dt_string = now.strftime("%d/%m/%Y %H:%M:%S")
print("date and time =", dt_string)
```
