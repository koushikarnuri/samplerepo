```python
# Identifying numerical columns in the dataset
num_columns = df.select_dtypes(include=['number']).columns
print(num_columns)
```

```python
# Calculate the mean, median, mode, and standard deviation for these columns.
df[num_columns].mean()
mean=print(df[num_columns].mean())
```

```python
df[num_columns].median()
median=print(df[num_columns].median())
```

```python
df[num_columns].mode()
mode=print(df[num_columns].mode())
```

```python
df[num_columns].std()
std=print(df[num_columns].std())
```

**Breif Interpretation** = he data looks like  heavily skewed to the right in price and sales-related columns, meaning a few high-value sales/products dominate the averages.
Median values better represent the “typical” transaction than the means.
Discounts are relatively stable in rate but vary widely in amount, depending on the transaction size.
Overall, the dataset suggests that typical transactions are modest in price, sales value, and discounts, but the presence of high-value sales significantly impacts the averages, making the median a more reliable measure of central tendency for understanding general sales patterns.

**DATA VISUALIZATION**

```python
# Histograms
df[num_columns].hist(figsize=(10, 6))
plt.tight_layout()
plt.show()
```

**Analysis and Inference** = The dataset is dominated by small to mid-range transactions, while a few high-value sales and expensive products introduce heavy right skew and outliers.

median values are more reliable than means since they better represent typical customer behavior.

The Right skewness and presence of outliers suggest different customer segments: regular small buyers vs. occasional bulk/high-value buyers.

```python
# Boxplot
df[num_columns].plot(kind='box', figsize=(10, 6))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Findings** = The dataset contains so many upper outliers in financial variables like price, sales, discounts reflecting premium or bulk items.
Volume also has noticable outliers, suggesting a skewed distribution with a few products selling in large quantities.
Discount rate outliers are mainly lower values, showing inconsistent discount strategies across products.

```python
# Bar Chart Analysis for Categorical Column
cat_columns = df.select_dtypes(include=['object']).columns
print(cat_columns)
```

```python
plt.figure(figsize=(15, 12))
for i, col in enumerate(cat_columns, 1):
   plt.subplot(len(cat_columns)//2+1,2,i)
   df[col].value_counts().plot(kind='bar', edgecolor='red')
   plt.title(f"Frequency of {col}")
   plt.xlabel(col)
   plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

**Key Findings** = The business primarily runs on many small, consistent transactions, but its averages are shaped by a few large, high-value sales. Median values give a more realistic view of everyday sales patterns. The dataset shows a long-tail pattern—most sales are modest, but a small number of products or transactions drive disproportionately high values in price, sales, and discounts. The dataset is dominated by small to mid-range transactions, while a few high-value sales and expensive products introduce heavy right skew and outliers. Sales are concentrated in specific days, cities, and product categories. The business is brand-driven, with only a few models/SKUs accounting for most sales. These findings can guide inventory planning, targeted promotions, and city-specific marketing strategies.
